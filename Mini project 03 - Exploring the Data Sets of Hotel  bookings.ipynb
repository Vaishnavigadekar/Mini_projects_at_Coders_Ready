## Conclusion:

1. According to given data, two types of hotels categories preferred by customers are "City Hotel" and "Resort Hotel".
2. Among these 2 categories "City Hotel" have more booking than "Resort Hotel" throughout the year.
3. August, July, May, October and April had highest number of bookings in year respectively.
4. Both categories have bookings mainly in mid-time of the month i,e around 15th  date of each month.
5. There are more average days in waiting list for "City Hotel" than "Resort Hotel".
6. On average Customer mainly prefers "Resort Hotel" for weekend night stays than "City Hotel".
7. During weekdays also on average customer prefers "Resort Hotel" than "City Hotel" for night stays.
